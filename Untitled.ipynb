```python
def preprocess_lines(file_path, expected_number_of_columns):
    with open(file_path, 'r') as f:
        cleaned_lines = [line.strip() for line in f.readlines() if len(line.split(',')) == 5]
    return '\n'.join(cleaned_lines)

file_path = 'spam.csv'
expected_number_of_columns = 5  # Replace with the actual expected number of columns
cleaned_data = preprocess_lines(file_path, expected_number_of_columns)

# Now use the cleaned data with read_csv
df = pd.read_csv(io.StringIO(cleaned_data))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
```

```python
#Data Preprocessing
def transform_text(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    y=[]
    for i in text:
        if i.isalnum():#we are removing % / and * or any other symbols@
            y.append(i)
    text=y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text=y[:]
    y.clear()   
    for i in text:
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps.stem('dancing')
transform_text('I loved the lectures on Machine learning How about you ')
```

```python
df['transformed_text']=df['text'].apply(transform_text)
```

# Model Building

```python
#It is considered that Naives Bayes works better on string
#String vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
x=cv.fit_transform(df['transformed_text']).toarray()
```

```python
x.shape
```

```python
y=df['target'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
classifier_1=GaussianNB()
classifier_2=MultinomialNB()
classifier_3=BernoulliNB()
```

```python
classifier_1.fit(x_train,y_train)
```

```python
y_pred=classifier_1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

```python
classifier_2.fit(x_train,y_train)
```

```python
y_pred=classifier_2.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

```python
classifier_3.fit(x_train,y_train)
```

```python
y_pred=classifier_3.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)#10 Trees
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(precision_score(y_test,y_pred))
```
